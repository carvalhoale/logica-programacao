##Condicional



```python
#if
if 5 < 10:
  print("Verdade")
```

```python
#if/else
num1 = 10
num2 = 15
if num1 > num2:
  print("O num1 é o maior número")
else:
  print("O num2 é o maior número")
```

```python
#if/elif/else
num1 = 10
num2 = 10
if num1 > num2:
    print( "O num1 é o maior número")
elif num1 < num2:
    print( "O num2 é o maoir número")
else:
      print ("Os dois são iguais")
```

##Condicional com operadores lógicos

```python
num1 = 10
num2 =5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print ("O num1 é o maior número")
else:
  print("O num1 não é o maior número")
```
